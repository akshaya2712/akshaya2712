#Data Preprocessing


## read data

```python
import pandas as pd
```

```python
df=pd.read_csv("/content/total_data.csv")
```

```python
df
```

```python
df.info()
```

## raw x,y

```python
unprocessed_x=df.drop(['Sleep Disorder','Occupation','Person ID'],axis=1)
```

```python
unprocessed_x
```

```python
unprocessed_y = df['Sleep Disorder']
```

```python
unprocessed_y
```

##data encoding

```python
from sklearn.preprocessing import LabelEncoder
```

```python
label_encoder=LabelEncoder()
```

### encoding  Y


```python
y=label_encoder.fit_transform(unprocessed_y)
```

```python
y=pd.DataFrame(y)
```

### encoding X

#### encoding useful columns

```python
bmi=label_encoder.fit_transform(unprocessed_x['BMI Category'])
```

```python
bmidf=pd.DataFrame(bmi)
```

```python
gx=label_encoder.fit_transform(unprocessed_x['Gender'])
```

```python
gxdf=pd.DataFrame(gx)
```

```python
x1=unprocessed_x.drop('Gender',axis=1)
```

```python
x1df=pd.DataFrame(x1)
```

```python
x=pd.concat([x1df,gxdf],axis=1)
```

```python
x=x.rename(columns={x.columns[-1]:'Gender'})
```

```python
x=pd.concat([x,bmidf],axis=1)
```

```python
x=x.drop('BMI Category',axis=1)
```

```python
x=x.rename(columns={x.columns[-1]:'BMI Category'})
```

```python
x[['Systolic BP', 'Diastolic BP']] = x['Blood Pressure'].str.split('/', expand=True)
```

```python
x['Systolic BP'] = x['Systolic BP'].astype(float)
```

```python
x['Diastolic BP'] = x['Diastolic BP'].astype(float)
```

```python
x=x.drop('Blood Pressure',axis=1)
```

```python
y
```

```python
x
```

##data split

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

# Model

## decision tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
model=DecisionTreeClassifier()
```

```python
model.fit(x_train,y_train)
```

#Testing Model

```python
y_pred=model.predict(x_test)
```

```python
y_test
```

```python
from sklearn import metrics
```

```python

print(metrics.accuracy_score(y_test,y_pred))
```
